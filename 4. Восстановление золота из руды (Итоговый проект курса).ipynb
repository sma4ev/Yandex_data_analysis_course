Данные находятся в трёх файлах:

### Проверим, что эффективность обогащения рассчитана правильно

#### Вычислим эффективность обогащения на обучающей выборке для признака rougher.output.recovery

```python
#рассчитаем recovery на обучающей выборке
#доля золота в концентрате после флотации/очистки
C = data_train['rougher.output.concentrate_au']
#доля золота в сырье/концентрате до флотации/очистки
F = data_train['rougher.input.feed_au']
#доля золота в отвальных хвостах после флотации/очистки
T = data_train['rougher.output.tail_au']
recovery_calc = (C*(F-T))/(F*(C-T))*100
recovery_calc.head()
```

Сравним со значениями в обучающей выборке.

```python
data_train['rougher.output.recovery'].head()
```

#### Найдем MAE между расчётами и значением признака.

```python
MAE=mean_absolute_error(recovery_calc, data_train['rougher.output.recovery'])
print('MAE между рассчетным значеием и значением признака:', MAE)
```

MAE ~ `0`. Рассчетное значение `recovery` совпадает со значением признака.

- Заполнение интерполяцией выглядит реалестичней, например, средним.
- Интерполяция не моделирует флуктуации.
- Возможно в ряде есть сезонность или сильная зависимость от других параметров. Если бы это было известно, то можно было бы смоделировать нужные значения. Однако для задачи заполнения пропусков достаточно и линейной интерполяции.

Заполним таким способом пропуски в данном поле в трейнеровочном и полном датасете.

```python
#полный датасет
data_full=data_full.interpolate(method='linear').copy()
#трейнеровочный датасет
data_train=data_train.interpolate(method='linear').copy()
```

Чтобы не искажать результаты тестов, удалим пропуски из тестировочного датасета.

```python
print('Размер тестовой выборки до удаления пропусков:', len(data_test))
print('Размер тестовой выборки после:', len(data_test.dropna()))
print('Потеряно: {:.2%}'.format((len(data_test)-len(data_test.dropna()))/len(data_test)))
#удалим пропуски
data_test=data_test.dropna()
```

Потеряна небольшая доля тестовой выборки, в пользу качества тестов.

### Напишем функцию для вычисления итоговой sMAPE.

Напишем функцию sMAPE.

```python
#sMape
def s_mape(y_fact, y_pred):
    y_fact = y_fact.to_numpy()
    return (abs(y_fact-y_pred)/((abs(y_fact)+abs(y_pred))/2)).mean()*100
    
#Итоговая sMape
def s_mape_final(routher, final):
    #если пришли float значения
    if type(routher)==np.float64 and type(routher)==np.float64:
        return (0.25*routher)+(0.75*final)
    else:
        return (0.25*s_mape(*routher))+(0.75*s_mape(*final))

#сделаем скорер
scorer = make_scorer(s_mape, greater_is_better=False)
```

#### Сравним результат с фиктивной моделью

Сравним результаты работы моделей с фиктивным регрессором - DummyRegressor. Будем использовать стратегии - `mean` и `median`. Прогноз регрессора сводится к среднему или медиане.

```python
for strategy in ('mean', 'median'):
    #создадим и обучим модель
    model_dr = DummyRegressor(strategy=strategy).fit(data_train_features, data_train_target)
    #получим прозноз                     
    model_dr_pred = model_dr.predict(data_test_features)
    #оценим sMape целевых по отдельности
    cross_val_dr_rougher = -cross_val_score(model_dr, data_train_features, data_train_target['rougher.output.recovery'], cv=5, scoring=scorer).mean()
    cross_val_dr_final = -cross_val_score(model_dr, data_train_features, data_train_target['final.output.recovery'], cv=5, scoring=scorer).mean()
    #посчитаем общий
    s_mape_dr = s_mape_final(cross_val_dr_rougher, cross_val_dr_final)
    print('Стратегия:', strategy, ' sMAPE:', s_mape_dr.round(3))
```
